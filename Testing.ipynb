## imports

```
from keras.preprocessing.image import img_to_array, load_img
from skimage.color import rgb2lab, lab2rgb
from skimage.transform import resize
from skimage.io import imsave
import numpy as np
import tensorflow as tf
import cv2
```

## configs

```
# set network input size
SIZE = 224
```

```
# set test images path
img_test_path = ".../datasets/test"
img_result_path = ".../datasets/results/"
```

```
# load model 
model = tf.keras.models.load_model(".../models/image_colorization.model", custom_objects=None, compile=True)
# model = tf.keras.models.load_model(".../models/image_colorization.hdf5", custom_objects=None, compile=True)
```

## testing

```
# load test images and save predicted colorized versions

for i, filename in enumerate(glob.glob(img_test_path+'*.jpg')):
  print("-- "+str(i)+"/"+str(enumerate(glob.glob(img_test_path+'*.jpg')))+" --")
  img_original=[]
  img_resized=[]

  image_orig = img_to_array(load_img(filename))
 
  w = image_orig.shape[0]
  h = image_orig.shape[1]
  
  img_original.append(image_orig)
  img_original = np.array(img_original, dtype=float)
  img_original = rgb2lab(1.0/255*img_original)[:,:,:,0]
  img_original = img_original.reshape(img_original.shape+(1,))
 
  image = image_orig
  image = resize(image ,(SIZE,SIZE))

  img_resized.append(image)
  img_resized = np.array(img_resized, dtype=float)
  img_resized = rgb2lab(1.0/255*img_resized)[:,:,:,0]
  img_resized = img_resized.reshape(img_resized.shape+(1,))

  predicted = model.predict(img_resized)


  predicted = predicted*128

  predicted = cv2.resize(predicted[0], (h, w))
  
  result = np.zeros((w, h, 3))

  result[:,:,0] = img_original[0][:,:,0]
  result[:,:,1:] = predicted

  result = lab2rgb(result)

  imsave(img_result_path+str(i)+".png", result)
```
